```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

def median_filter(image, kernel_size=3):
    # Apply the median filter using a sliding window approach
    return cv2.medianBlur(image, kernel_size)

# Read an image (ensure the image path is correct)
image = cv2.imread('/content/URMILA.webp', cv2.IMREAD_GRAYSCALE)

# Apply the median filter with a 3x3 kernel
filtered_image = median_filter(image, kernel_size=3)

# Display original and filtered images
plt.figure(figsize=(10, 5))

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(image, cmap='gray')
plt.title('Original Image')
plt.axis('off')

# Filtered Image
plt.subplot(1, 2, 2)
plt.imshow(filtered_image, cmap='gray')
plt.title('Median Filtered Image')
plt.axis('off')

plt.show()
```
